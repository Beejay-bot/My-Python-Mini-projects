SIMULATING STOCK DATA

```python
prices = np.full(100, fill_value=(np.nan))
```

```python
prices
```

```python
prices[[0, 30, 80, -1 ]] = [75, 30, 80, 60]
prices
```

```python
length_list = np.arange(len(prices))
is_valid = ~np.isnan(prices)#test for not a number values
is_valid 
```

```python
prices[is_valid]
```

```python
prices = np.interp(x=length_list, xp =length_list[is_valid], fp = prices[is_valid])
prices
```

```python
prices += np.random.rand(len(prices))*2
prices
```

```python
mn = np.argmin(prices)
mx = mn + np.argmax(prices[mn:])
kwargs = {'markersize': 12, 'linestyle': ''}
```

```python
import matplotlib.pyplot as plt

plt.plot(prices)
plt.title('Price History')
plt.xlabel('Time')
plt.ylabel('Price')
plt.scatter(mn, prices[mn], color='red')
plt.scatter(mx, prices[mx], color='green')
plt.grid(b=None, which='major', axis='both')
plt.show()
```
