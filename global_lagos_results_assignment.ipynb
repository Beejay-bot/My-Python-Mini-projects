```python
import matplotlib.pyplot as plt 
import random
import seaborn as sns
import pandas
import numpy as np 
from math import pi
from numpy import sin,cos
plt.style.use("ggplot")
```

```python
global_results = pandas.read_csv('global_results.csv')
lagos = pandas.read_csv('lagos_results.csv')
```

```python
# lagos.columns
# global_results.columns
```

```python
values_missing_avgtemp = round(lagos["avg_temp"].mean(), 2)
lagos["avg_temp"].fillna(values_missing_avgtemp,inplace = True)
# lagos.info()
```

```python
# average_of_global_results = global_results.groupby("avg_temp").mean().reset_index()
# average_of_global_results
average_of_lagos_results = lagos.groupby("avg_temp").mean().reset_index()
lagos
```

```python
lagos["moving_avg"] = lagos.avg_temp.rolling(window = 5).mean()
global_results["moving_avg"] = global_results.avg_temp.rolling(window = 5).mean()
```

```python
fig,axs = plt.subplots(1,2, figsize = (20,5))
fig.suptitle('Results')
plot1 = axs[0]
plot2 = axs[1]

plot1.set_title("Lagos")
plot2.set_title("Global")



x_lag=lagos['year']
y_lag=lagos['avg_temp']


x_glo=global_results['year']
y_glo=global_results['avg_temp']
 


plot1.plot(x_lag, y_lag, color = "blue", alpha = 0.7)
plot1.plot(x_lag, lagos.moving_avg, color = "red", alpha = 0.7)
plot2.plot(x_glo, y_glo, color = "cyan", alpha = 0.7)
plot2.plot(x_glo, global_results.moving_avg, color = "red", alpha = 0.7)

plot1.set_xlabel('YEAR')
plot1.set_ylabel('TEMP')

plot2.set_xlabel('YEAR')
plot2.set_ylabel('TEMP')

plt.show()
```

## FILL BETWEEN

```python
plt.fill_between(lagos['year'], lagos['moving_avg'], alpha = 0.5, color = "yellow", label = "Lagos results moving average")
plt.fill_between(global_results['year'], global_results['moving_avg'], alpha = 0.5, color = "green", label = "Global results moving average")
plt.legend()
```
