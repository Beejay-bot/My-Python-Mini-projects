```python
import matplotlib.pyplot as plt 
import random
import seaborn as sns
import pandas
import numpy as np 
from math import pi
from numpy import sin,cos
from mpl_toolkits.mplot3d import Axes3D 

plt.style.use("ggplot")
```

```python
# global_results = pandas.read_csv('global_results.csv')
ghana = pandas.read_csv('ghana.csv')
ghs_usd = pandas.read_csv('GHS_USD Historical Data.csv')
ghana_details = pandas.read_csv('GHS_USD Historical Data (1).csv')
nigeria = pandas.read_csv('nigeria.csv')
```

```python
fig,axs = plt.subplots(1,2, figsize = (15,5))
fig.suptitle('GHANA')
plot1 = axs[0]
plot2 = axs[1]

plot1.set_title("year X population")
plot2.set_title("year X unemployment")

x_pop=ghana['year']
y_pop=ghana['population_m']

x_un=ghana['year']
y_un=ghana['unemployment%']

plot1.barh(x_pop, y_pop, color = "blue", alpha = 0.4)
plot2.barh(x_un, y_un, color = "red", alpha = 0.4)
# plot2.plot(x_glo, global_results.moving_avg, color = "red", alpha = 0.7)

plot1.set_xlabel('YEAR')
plot1.set_ylabel('POPULATION')

plot2.set_xlabel('YEAR')
plot2.set_ylabel('POPULATION')

plt.show()
```

```python
Budget = round(ghana_details["Budget"].mean())
ghana_details["Budget"].fillna(Budget,inplace = True)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x =ghana_details['population']
y =ghana_details['year']
z =ghana_details['Budget']

ax.plot(x,y,z, c = 'r', marker = 'o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('z Label')
plt.show()
```
