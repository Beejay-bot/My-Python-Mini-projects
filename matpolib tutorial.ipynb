```python
t = np.linspace(0,2, num=300)
```

```python
sq = 0
sw = 0
f = 1
for k in range(1,6):
    sq = sq + ((np.sin(2*pi*((2*k)-1)*t)/((2*k)-1)))
    sw = sw + ((np.sin(2*pi*((2*k)-1)*t)/((2*k)-1)))
    s_sum = (2/pi)*sw
    sum_sq = (4/pi)*sq
    plt.plot(t,sum_sq)
    plt.plot(t, sw)
    plt.title('Square wave')
    plt.grid(True, which = 'both')
    plt.plot(sq,sw, color = 'blue', marker = '+', linestyle = '-', markersize = 8)
    plt.plot(sq, sw, color = 'red', marker = '*', linestyle = '-.', markersize = 8)
```

```python
x = np.linspace(-pi,pi, 100)
ysin = sin(x)
ycos = cos(x)

def create_plot(c,v,b,n,m):
    plt.plot(c,v)
    plt.plot(c,b)
    plt.ylabel("y")
    plt.xlabel("x")
    plt.legend((n,m))
    plt.title('sin (x) and cos(x) from -pi to pi')
create_plot(x,ysin,ycos, 'sin(x)', 'cos(x)')
```
